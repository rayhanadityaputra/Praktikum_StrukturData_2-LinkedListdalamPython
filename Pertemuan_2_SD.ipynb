**2.4. MENGHAPUS NODE PADA LINKED LIST**

```python
# HAPUS NODE DENGAN  KATA KUNCI

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

#Function to add node in the beginning
  def AddBeginning(self, newdata):
    NewNode = Node(newdata)
# updates the new nodes next val to existing node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Function to remove node by key
  def RemoveNode(self, Removekey):
    HeadVal = self.headval
    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.next
        HeadVal = None
        return
    while (HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval
    if (HeadVal == None):
      return
    prev.nextval = HeadVal.nextval
    HeadVal = None

# create linked list nama Li
Li = LinkedList()

# add node
Li.AddBeginning("Januari")
Li.AddBeginning("Februari")
Li.AddBeginning("Maret")
Li.AddBeginning("April")
print("Sebelum dihapus")
Li.listprint()
print("===================================================")

# remove node by key == "Maret"
Li.RemoveNode("Februari")
print("setelah dihapus dengan kunci 'Maret'")
Li.listprint()
```

```python
# HAPUS NODE DI AWAL

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

#Function to add node in the end
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval = NewNode

# Function to remove last node
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

# create linked list nama Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
Li.AddInEnd("April")
print("Sebelum dihapus")
Li.listprint()
print("===================================================")

# remove last node
Li.RemoveEnd()
print("setelah dihapus node pertama")
Li.listprint()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Tambah di akhir
    def insert_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    # Sisip di tengah (setelah node pertama)
    def insert_middle(self, data):
        if self.head is None:
            return

        new_node = Node(data)
        temp = self.head.next

        self.head.next = new_node
        new_node.prev = self.head
        new_node.next = temp

        if temp:
            temp.prev = new_node

    # Hapus berdasarkan key
    def delete_key(self, key):
        temp = self.head

        while temp and temp.data != key:
            temp = temp.next

        if temp is None:
            print("Data tidak ditemukan")
            return

        if temp.prev:
            temp.prev.next = temp.next
        else:
            self.head = temp.next

        if temp.next:
            temp.next.prev = temp.prev

    # Tampilkan
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")


# === MAIN PROGRAM ===
dll = DoublyLinkedList()

# Buat 3 node
dll.insert_end(10)
dll.insert_end(20)
dll.insert_end(30)

print("Awal:")
dll.display()

# Sisip tengah
dll.insert_middle(15)
print("Setelah sisip tengah:")
dll.display()

# Hapus key 20
dll.delete_key(20)
print("Setelah hapus 20:")
dll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Insert di akhir
    def insert_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    # Hapus node akhir
    def delete_end(self):
        if self.head is None:
            return

        temp = self.head

        # Jika hanya satu node
        if temp.next == self.head:
            self.head = None
            return

        prev = None
        while temp.next != self.head:
            prev = temp
            temp = temp.next

        prev.next = self.head

    # Tampilkan
    def display(self):
        if self.head is None:
            return

        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(kembali ke head)")


# === MAIN PROGRAM ===
cll = CircularLinkedList()

# Buat 3 node
cll.insert_end(1)
cll.insert_end(2)
cll.insert_end(3)

print("\nAwal:")
cll.display()

# Sisip akhir
cll.insert_end(4)
print("Setelah sisip akhir:")
cll.display()

# Hapus akhir
cll.delete_end()
print("Setelah hapus akhir:")
cll.display()
```
